```python
import pandas as pd
```

```python
df = pd.read_csv("shl_scraped_catalog.csv", encoding="utf-8")

print(df.shape)
print(list(df.columns))

display(df.head(3))
```

```python
df = df.rename(columns={
    "assessment_name": "name",
    "remote_testing": "remote_support",
    "adaptive_testing": "adaptive_support",
    "test_type": "test_type",          
    "description": "description",
    "duration": "duration",
    "job_levels": "job_levels",
    "url": "url"
})
```

```python
print(df.isnull().sum())

print("\nüîÅ Duplicate Rows:", df.duplicated().sum())
```

```python
df["duration"] = df["duration"].fillna("N/A")
df["job_levels"] = df["job_levels"].fillna("N/A")


print("‚úÖ Missing values fixed!")
print(df.isnull().sum())
```

```python
if "url" in df.columns:
    print("üîÅ Duplicate URLs:", df["url"].duplicated().sum())
```

```python
df.to_csv("shl_catalog_clean.csv", index=False, encoding="utf-8")
print(" Saved cleaned dataset as shl_catalog_clean.csv")
```
